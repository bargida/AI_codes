```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import math
%matplotlib inline
```

```python
def load_my_data():
  data = np.loadtxt('ex1data1.txt', delimiter=',')
  input_features = data[:,0]
  target_outputs = data[:,1]

  return (input_features, target_outputs)
```

```python
x_train, y_train = load_my_data()

print('x_train', x_train)
print('\n')
print('y_train', y_train)
```

```python
def overall_error(x, y, w, b):

  m = len(x)

  cost_sum = 0

  for i in range(m):

    f_wb = w * x[i] + b
    cost_i = (f_wb - y[i]) ** 2
    cost_sum = cost_sum + cost_i

  overall_model_average_error = cost_sum / (2 * m)

  return overall_model_average_error
```

```python
average_model_error = overall_error(x_train, y_train, 2, 1)
print('average_model_error :' , average_model_error)
```

```python
def calculate_gradients(x, y, w, b):

    m = len(y)

    dj_db = 0

    dj_dw = 0

    for i in range(m):

      f_wb = w * x[i] +b

      dj_db_i = f_wb - y[i]

      dj_dw_i = (f_wb - y[i]) * x[i]

      dj_db = dj_db +dj_db_i

      dj_dw += dj_dw_i

    return(dj_db /m, dj_dw /m)
```

```python
print('dj_db = ', calculate_gradients(x_train, y_train, 2, 1)[0])
print('dj_dw = ', calculate_gradients(x_train, y_train, 2, 1)[1])
```

```python
def gradient_descent_algorithm(x, y, w_in, b_in, overall_error, calculate_gradients, alpha, num_iter):

  m = len(x)
  w_history_values = []
  j_history_values = []


  b = b_in
  w = w_in

  for i in range(num_iter):

    dj_db, dj_dw = calculate_gradients(x, y, w, b)

    w = w - alpha * dj_dw

    b = b - alpha * dj_db

    if i % math.ceil(num_iter / 10) == 0:
      w_history_values.append(w)
      j_history_values.append(overall_error(x, y, w, b))

  return w, b, w_history_values, j_history_values

w_final, b_final, w_historic_values,  j_historic_values = gradient_descent_algorithm(x_train, y_train, 0, 0, overall_error, calculate_gradients, 0.02, 100)

print('Final value for w', w_final)
print('Final value for w', b_final)

```

```python
predicted_target_output_values = np.zeros(len(x_train))

for i in range(len(x_train)):

  predicted_target_output_values[i] = w_final * x_train[i] + b_final
```

```python

def model_predict_function(x, w, b):
  m = x.shape[0]
  n = x.shape[1]

  f_predicted_outputs = np.zeros(m)

  for i in range(m):
    predicted = np.dot(x[i, :], w) + b
    f_predicted_outputs[i] = predicted

  return f_predicted_outputs
```

```python
predicted_target_values = model_predict_function(x_train, w_initial, b_initial)
print("The actual target values are: ", y_train)
print("The predicted target values are: ", predicted_target_values)
```

```python
def compute_error_cost(x, y, w, b):

    m = x.shape[0]

    cost = 0

    for i in range(m):

        f_wb_i = np.dot(x[i, :], w) + b
        cost = cost + (f_wb_i - y[i]) ** 2

    cost = cost / (2 * m)

    return cost
```

```python
print('Overall average error : ', compute_error_cost(x_train, y_train, w_initial, b_initial))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('earthquakes.csv', index_col=0, parse_dates=[0], usecols=[0, 1])
#parse_dates bu df ichidan (2010-01-20) aynan bittasini ajratib olish u/n ishlatamiz Masalan:

# df['year] = [i.year for i in df.index] (yilini ajratib u/n)
#df.head

df.reset_index(inplace=True) #duplicate bob qogani uchun reindexladim
df = df[~df.index.duplicated(keep='first')]
df.drop(['year', 'day', 'month', 'richter', 'area'], axis=1, inplace=True, errors='ignore')
df.columns = ['region', 'deaths']
df
```

```python
df.shape
```

```python
plt.figure(figsize=(30,8))
sns.barplot(x=df.index, y=df['deaths'])
plt.xticks(rotation=90) #tik larni orasini ochib beradi
plt.show()
```

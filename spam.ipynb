```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
df['Category'].value_counts()
```

```python
df['spam_or_no_spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message,df.spam_or_no_spam,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
type(x_train)
```

```python
# Converting X_train and x_test into sparse matrices:

from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:2]
```

```python
x_train_np = x_train_count.toarray()

np.where(x_train_np[1] != 0)
```

```python
#Training the model:

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(x_train_count,y_train)
```

```python
#After training the model, we can evaluate the performance of the model with testing subset:

from sklearn.metrics import classification_report

y_pred = model.predict(v.transform(x_test))

print(classification_report(y_test,y_pred))
```

```python
brand_new_emails = ['Hey Andrew, will we go to swimming pool tomorrow?',
                    "Wow amazing, you are the winner of $ 5 million. Claim it immediately by clicking the following link"]


brand_new_emails_count = v.transform(brand_new_emails)

model.predict(brand_new_emails_count)

print(model.predict(brand_new_emails_count))
```

```python
#Combining gCountvectorizer and model training steps into one single pipeline:
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])

clf.fit(x_train, y_train)
```

```python
#Evaluating the model's accuracy:

y_pred = clf.predict(x_test)

print(classification_report(y_test,y_pred))
```

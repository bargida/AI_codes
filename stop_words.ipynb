```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS

STOP_WORDS
print(len (STOP_WORDS))
```

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp("There are lots of trees in this island and many of them are 100-year-old.")

for word_token in doc:
  if word_token.is_stop:
    print(word_token, 'is  a stop word')
  else:
    print(word_token, 'is not a stop word')

```

```python
#Creating a custom to separate words from a given text which are NOT STOP WORDS:

def remove_stop_words(text):
    doc = nlp(text)

    no_stop_words = []

    for word_token in doc:
        if not word_token.is_stop:
            no_stop_words.append(word_token.text)

    return ' '.join(no_stop_words)
```

```python
#Let's check our custom function:
print(remove_stop_words('In this sentence there are lots of STOP WORDS. And all those words should be removed !'))
```

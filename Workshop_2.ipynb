* Exercise 2: Write a function that receives the dataset and replace Famle with F and Male with M (please try to write it yourself before checking the answer in the next cell)

```python
# exercise 2 solution
def encode_sex(data):
    data.reset_index(drop=True, inplace=True) # reindexing the dataset in case the dataset index is corrupted
    rows=data.shape[0]
    a=data.loc[:,'sex']
    for i in range(rows):
        if a[i]=="Male":
            data.loc[i,"sex"]="M"
        elif a[i]=="Female":
            data.loc[i,"sex"]="F"
    return data
```

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)  ####################
import pandas as pd
data = pd.read_csv('adult.csv')
str_cols = data.select_dtypes(include='object').columns
data[str_cols] = data[str_cols].apply(lambda c: c.str.strip())

# randomly 10000 samples
data_new = data.sample(n=10000, random_state=48)

# drop 'income', clean
data_new = data_new.drop(['income'], axis=1)
data_new.reset_index(drop=True, inplace=True)
data_new = data_new.replace('?', pd.NA).dropna()
data_new.reset_index(drop=True, inplace=True)

# print data
print(data_new.shape)
data_new
```

## 3- Normalise the numerical columns.

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)  ####################
# from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data_f = data_new.copy()

# identify categorical and numerical columns
categorical_cols = data_f.select_dtypes(include='object').columns.tolist()
numerical_cols   = data_f.select_dtypes(include='number').columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numerical columns:  ", numerical_cols)

# encode categorical columns using Label Encoding
le = LabelEncoder()
for col in categorical_cols:
    data_f[col] = le.fit_transform(data_f[col])

# normalise numerical columns using Min-Max scaling (0 to 1)
scaler = MinMaxScaler()
data_f[numerical_cols] = scaler.fit_transform(data_f[numerical_cols])

# print
print(data_f)
```

## P 2.3: Write a function to split the dataset in half column-wise and swap the first half and the second half. Apply the funtion to the dataset and print the result (3%)

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)  ####################
def swap_column_halves(df):
    cols = df.columns.tolist()
    mid = len(cols) // 2
    first_half = cols[:mid]
    second_half = cols[mid:]
    return df[second_half + first_half]

data_ = swap_column_halves(data_f.copy())
print(data_)

```

## P 2.4: Write a function that receives two numerical columns' names and compare their values for all rows. If the value of the first column is greater than the second column, the function should produce True, otherwise, it should produce False. The function should append an additional column to the dataset to store the results of the comparison for all rows. Apply the function to the "age" and "hours-per-week" columns in the dataset and print the result (4%).

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)  ####################
def compare_columns(df, col1, col2):
    result_col = f'{col1}_greater_than_{col2}'
    df[result_col] = df[col1] > df[col2]
    return df

data_2 = compare_columns(data_f.copy(), 'age', 'hours-per-week')
print(data_2)

```
